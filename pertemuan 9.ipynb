```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('titanic.csv')

# ---------------------------------------------------------
# SOAL NO 1: Identifikasi rata-rata usia, fare, survive, pclass
# ---------------------------------------------------------
mean_age = df['age'].mean()
mean_fare = df['fare'].mean()
mean_survive = df['survived'].mean()
mean_pclass = df['pclass'].mean()

print("1. IDENTIFIKASI RATA-RATA")
print(f"   - Rata-rata Usia (Age)     : {mean_age:.2f} tahun")
print(f"   - Rata-rata Tarif (Fare)   : {mean_fare:.2f}")
print(f"   - Rata-rata Survive        : {mean_survive:.2f}")
print(f"   - Rata-rata Pclass         : {mean_pclass:.2f}")
print("-" * 40)
print()

# ---------------------------------------------------------
# SOAL NO 2: Identifikasi max dan min usia dan fare
# ---------------------------------------------------------
max_age = df['age'].max()
min_age = df['age'].min()

max_fare = df['fare'].max()
min_fare = df['fare'].min()

print("2. IDENTIFIKASI MAX DAN MIN")
print(f"   - Usia (Age) -> Max: {max_age} tahun, Min: {min_age} tahun")
print(f"   - Tarif (Fare)-> Max: {max_fare}, Min: {min_fare}")
print("-" * 40)
print()

# ---------------------------------------------------------
# SOAL NO 3: Identifikasi outlier fare (Record Data & Visual)
# ---------------------------------------------------------
print("3. IDENTIFIKASI OUTLIER FARE")

Q1 = df['fare'].quantile(0.25)
Q3 = df['fare'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

print(f"(Perhitungan Batas Atas Outlier: {upper_bound:.2f})")

outlier_data = df[df['fare'] > upper_bound]
print(f"Ditemukan {len(outlier_data)} data outlier.")
print("Berikut 5 contoh data outlier teratas:")
print(outlier_data[['name', 'fare', 'pclass']].head(5))

print("\n   [Menampilkan Visualisasi Boxplot...]")

plt.figure(figsize=(8, 6))
plt.boxplot(df['fare'])
plt.title("Box Plot Outlier Tarif (Fare) Titanic")
plt.ylabel("Fare")
plt.grid(True, alpha=0.3)
plt.show()

print("\n   [Menampilkan Visualisasi Histogram...]")

plt.figure(figsize=(12, 5))

# Plot 1: Histogram Tarif (Fare)
plt.subplot(1, 2, 1)
plt.hist(df['fare'], bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram Distribusi Tarif (Fare)")
plt.xlabel("Fare")
plt.ylabel("Frekuensi")
plt.grid(axis='y', alpha=0.5)

# Plot 2: Histogram Usia (Age))
plt.subplot(1, 2, 2)
plt.hist(df['age'].dropna(), bins=20, color='lightgreen', edgecolor='black')
plt.title("Histogram Distribusi Usia (Age)")
plt.xlabel("Age")
plt.ylabel("Frekuensi")
plt.grid(axis='y', alpha=0.5)

plt.tight_layout()
plt.show()
```
